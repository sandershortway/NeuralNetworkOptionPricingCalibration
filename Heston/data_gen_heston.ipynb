```python
import pandas as pd
import numpy as np
import heston as h

option = pd.DataFrame(columns=["r", "v0", "vbar", "kappa", "zeta", "rho", "maturity", "moneyness", "price"])

size = 25
numbers_of_parameters = 1000
```

```python
def shuffle_heston():
    param = {}
    param["s0"] = 100
    param["r"] = np.random.uniform(0, 0.15)
    param["v0"] = np.random.uniform(0, 0.3)
    param["vbar"] = np.random.uniform(0, 0.3)
    param["kappa"] = np.random.uniform(0, 1.5)
    param["zeta"] = np.random.uniform(0, 1.0)
    param["rho"] = np.random.uniform(-1, 0)
    return param
    
```

```python
T = np.linspace(1, 3, size)
moneyness = np.linspace(0.5, 1.5, size)
```

```python
for param_num in range(numbers_of_parameters):
    param = shuffle_heston()
    if param_num % 10 == 0:
        print(param_num, "/", numbers_of_parameters)
    for i in range(len(T)):
        for j in range(len(moneyness)):
            row = {
                "r": param["r"], 
                "v0": param["v0"], 
                "vbar": param["vbar"], 
                "kappa": param["kappa"], 
                "zeta": param["zeta"], 
                "rho": param["rho"], 
                "maturity": T[i], 
                "moneyness": moneyness[j], 
                "price": h.heston_euro_call_param(param, T[i], moneyness[j])
            }
            row_df = pd.DataFrame(row, index=[0])
            option = pd.concat([option, row_df], ignore_index=True)
            #option = option.append(row, ignore_index=True)
```
