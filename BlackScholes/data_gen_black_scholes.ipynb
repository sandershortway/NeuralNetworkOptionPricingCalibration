```python
import pandas as pd
import numpy as np
import black_scholes as bs

option = pd.DataFrame(columns=["r", "sigma", "maturity", "moneyness", "price"])

size = 25
numbers_of_parameters = 1000
```

```python
def shuffle_black_scholes():
    param = {}
    param["s0"] = 100
    param["r"] = np.random.uniform(0, 0.15)
    param["sigma"] = np.random.uniform(0, 0.3)
    return param
    
```

```python
T = np.linspace(1, 3, size)
moneyness = np.linspace(0.5, 1.5, size)
```

```python
for param_num in range(numbers_of_parameters):
    param = shuffle_black_scholes()
    if param_num % 10 == 0:
        print(param_num, "/", numbers_of_parameters)
    for i in range(len(T)):
        for j in range(len(moneyness)):
            row = {
                "r": param["r"], 
                "sigma": param["sigma"],
                "moneyness": moneyness[j],
                "maturity": T[i], 
                "price": bs.euro_call(param["s0"], moneyness[j] * param["s0"], T[i], param["r"], param["sigma"])
            }
            row_df = pd.DataFrame(row, index=[0])
            option = pd.concat([option, row_df], ignore_index=True)
            #option = option.append(row, ignore_index=True)
```
